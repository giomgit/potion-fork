```python
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
import pandas as pd
```

```python
log_dir = '../logs'
log_name = 'REINFORCE'
os.chdir(log_dir) #Change directory 
runs = [pd.read_csv(file, index_col=False) 
          for file in glob.glob("*.csv") if file.startswith(log_name + '_')]
```

```python
run = runs[0]
performance = run['Perf']
performance = performance[32:80] - 200
x = range(len(performance))
plt.plot(x, performance, 'none')
plt.xlabel('Policy updates')
plt.ylabel('Performance')

threshold = -20
plt.plot(x, np.zeros_like(x) + threshold, 'r--')

optim = 308
plt.plot(x, np.zeros_like(x) + optim, 'k--')

baseline = 11.3
rate = 0.05
plt.plot(x, baseline + (optim-baseline)*(1 - np.exp(-rate*np.array(x))), 'none')

plt.plot(0,baseline,'o', ms='12')

#plt.text(10, -100, 'DANGER!!', color='r', fontsize=14)

plt.rc('lines', linewidth=3)
plt.ylim(-150,350)
plt.savefig('imp1.pdf')
```
